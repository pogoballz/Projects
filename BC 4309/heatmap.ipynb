```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv("listing(ori).csv")
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),annot = True)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data.longitude,data.latitude,c= data.availability_365,cmap='bwr',edgecolor='black')
cbar = plt.colorbar()
cbar.set_label('availability_365')
```

```python
 df = data[data['price']<=1000]
plt.figure(figsize=(10,6))
plt.scatter(df.reviews_per_month,df.price,edgecolor='black')
plt.xlabel('review per month')
plt.ylabel('price')
```

```python
import numpy as np
def pearson_r(x,y):
    
    corr_mat = np.corrcoef(x, y)

    # Return entry [0,1]
    print(corr_mat[0,1])

r = pearson_r(data.reviews_per_month, data.price)
```
